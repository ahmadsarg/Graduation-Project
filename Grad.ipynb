````python
%%writefile app.py

import streamlit as st
import streamlit.components.v1 as components
from transformers import pipeline
import torch
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
)
import re
import pandas as pd
from git import Repo, GitCommandError
import os
import shutil
from bs4 import BeautifulSoup
from langchain.llms import HuggingFacePipeline
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
from langchain.vectorstores import Chroma
from langchain.document_loaders.generic import GenericLoader
from langchain.document_loaders.parsers import LanguageParser
from langchain.text_splitter import Language

st.set_page_config(
    page_title="Code Documentation and Analysis Tool",
    page_icon="💻",
    layout="wide",
    initial_sidebar_state="expanded",
)

# Model configuration
base_model = "codellama/CodeLlama-7b-Instruct-hf"

# Check if a GPU is available
if torch.cuda.is_available():
    compute_dtype = getattr(torch, "float16")

    quant_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=compute_dtype,
        bnb_4bit_use_double_quant=False,
    )
else:
    quant_config = None  # No quantization if no GPU is available

# Load the model and tokenizer
@st.cache_resource
def load_model():
    if torch.cuda.is_available():
        model = AutoModelForCausalLM.from_pretrained(
            base_model,
            quantization_config=quant_config,
            device_map="auto"
        )
    else:
        model = AutoModelForCausalLM.from_pretrained(base_model)

    model.config.use_cache = False
    model.config.pretraining_tp = 1

    tokenizer = AutoTokenizer.from_pretrained(base_model, trust_remote_code=True)
    tokenizer.pad_token = tokenizer.eos_token
    tokenizer.padding_side = "right"

    return model, tokenizer

# Create the Streamlit interface
model, tokenizer = load_model()

def render_mermaid(code: str, height: int = 500) -> None:
    """
    Renders a Mermaid diagram in the Streamlit app.
    """
    html_code = f"""
        <div class="mermaid" style="height: {height}px; overflow-y: auto;">
            {code}
        </div>
        <script src="https://cdn.jsdelivr.net/npm/mermaid@10/dist/mermaid.min.js"></script>
        <script>mermaid.initialize({{startOnLoad: true}});</script>
    """
    components.html(html_code, height=height)

def extract_mermaid_code(text):
    """
    Extracts the Mermaid code from a string using regex.

    Args:
        text: The text containing the Mermaid code block.

    Returns:
        The extracted Mermaid code as a string, or None if no code is found.
    """
    pattern = r"(?s)^.*?```mermaid\n((?:.|\n)*?)```$"

    match = re.findall(pattern, text, re.MULTILINE)

    if match:
        return match[0].strip()
    else:
        return None

def extract_functions_from_code(code):
    """
    Extracts function definitions from the given code.

    Args:
        code: The code as a string.

    Returns:
        A string containing the function definitions.
    """
    function_pattern = r"def\s+\w+\(.*?\):"
    functions = re.findall(function_pattern, code)
    return "\n".join(functions)

def extract_important_lines_from_html(html_content):
    """
    Extracts important lines from the given HTML content.

    Args:
        html_content: The HTML content as a string.

    Returns:
        A string containing the important lines from the HTML.
    """
    soup = BeautifulSoup(html_content, "html.parser")
    important_lines = []

    for tag in soup.find_all(["title", "h1", "h2", "h3", "p", "a"]):
        important_lines.append(tag.get_text(strip=True))

    return "\n".join(important_lines)


def document_files_page():
    st.header("📄 Document Files")
    st.markdown("Upload and analyze your code files.")
    uploaded_file = st.file_uploader("Choose a file", type=["py", "java", "cpp", "js", "sql", "html", "css", "kt", "php", "rb", "swift"])
    if uploaded_file:
        file_details = {"filename": uploaded_file.name, "filetype": uploaded_file.type, "filesize": uploaded_file.size}
        st.write(file_details)

        code = uploaded_file.read().decode("utf-8")
        file_extension = uploaded_file.name.split('.')[-1]
        st.code(code, language=file_extension)

        if file_extension == "html":
            # Extract and display important lines from HTML
            important_lines = extract_important_lines_from_html(code)
            st.write("Extracted Important Lines:")
            st.code(important_lines, language='html')

            # Generate explanation of the HTML content
            pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_new_tokens=4000, return_full_text=False)
            explanation_prompt = f"[INST] Analyze the following HTML content and provide a detailed explanation of what the web page is about:\n{important_lines}[/INST]"

            try:
                explanation = pipe(explanation_prompt)[0]["generated_text"]
            except Exception as e:
                explanation = f"Error generating explanation: {str(e)}"

            st.write("Generated Explanation:")
            st.write(explanation)

        else:
            # Extract and display functions for other code files
            functions = extract_functions_from_code(code)
            st.write("Extracted Functions:")
            st.code(functions, language=file_extension)

            # Step 1: Generate explanation of the code
            pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_new_tokens=4000, return_full_text=False)
            explanation_prompt = f"[INST] Analyze the following {file_extension} code and provide a detailed documentation for each function:\n{code}[/INST]"

            try:
                explanation = pipe(explanation_prompt)[0]["generated_text"]
            except Exception as e:
                explanation = f"Error generating explanation: {str(e)}"

            st.write("Generated Documentation:")
            st.write(explanation)

            # Step 2: Explain important functions
            function_explanation_prompt = f"[INST]Explain the most important functions from the following {file_extension} code:\n{functions}[/INST]"

            try:
                function_explanation = pipe(function_explanation_prompt)[0]["generated_text"]
            except Exception as e:
                function_explanation = f"Error generating function explanations: {str(e)}"

            st.write("Explanation of Important Functions:")
            st.write(function_explanation)

def home_page():
    st.header("🏠 Home")
    st.markdown("Generate documentation and mermaid diagrams from code prompts.")
    prompt = st.text_area("Enter your code prompt:", height=300, placeholder="Enter your code here...")

    if st.button("Generate Documentation and Mermaid Diagram"):
        # Step 1: Generate explanation of the code
        pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_new_tokens=5000, return_full_text=False, temperature=0.1, do_sample=True)
        explanation = pipe(prompt)[0]["generated_text"]
        st.write("Generated Explanation:")
        st.write(explanation)

        # Step 2: Generate Mermaid code from explanation
        mermaid_prompt = f"<s>[INST] <<SYS>>\\n you are a helpful coding assistant your task is to write mermaid diagram code for class diagrams from a given description\\n<</SYS>>\\n\\nWrite Mermaid.js class diagram code from the following description, make sure to include relations if any. Don't write any other explanation, only the mermaid code:\n{explanation} [/INST]"
        mermaid_output = pipe(mermaid_prompt)[0]["generated_text"]
        mermaid_code = extract_mermaid_code(mermaid_output)
        st.write("Generated Mermaid Code:")
        st.code(mermaid_prompt)
        st.write("Generated Mermaid Code:")
        st.code(mermaid_output)
        st.write("Extracted Mermaid Code:")
        st.code(mermaid_code)

        # Step 3: Render the Mermaid diagram
        st.write("Mermaid Diagram:")
        render_mermaid(mermaid_code, height=800)  # Adjust the height as needed

    if st.button("Add Comments to Code"):
        comment_prompt = f"[INST] <<SYS>>\\n you are a helpful coding assistant your task is to add inline comments and docstrings to the code provided.\\n<</SYS>>\\n\\n{prompt}[/INST]"
        comment_output = pipe(comment_prompt)[0]["generated_text"]
        st.write("Generated Comments:")
        st.code(comment_output)

# Define the pipeline and HuggingFace model
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_new_tokens=5000, return_full_text=False, temperature=0.1, do_sample=True)
llm = HuggingFacePipeline(pipeline=pipe)

def process_github_repo(github_url):
    repo_path = "/content/test_repo"  # Use a temp directory to clone the repo
    if os.path.exists(repo_path):
        shutil.rmtree(repo_path)

    try:
        repo = Repo.clone_from(github_url, to_path=repo_path)
    except GitCommandError as e:
        return str(e)

    loader = GenericLoader.from_filesystem(
        repo_path,
        glob="*/",
        suffixes=[".py", ".java", ".cpp", ".js", ".kt", ".php", ".rb", ".swift"],
        exclude=["/non-utf8-encoding.py"],
        parser=LanguageParser(language=Language.MULTILANGUAGE, parser_threshold=500),
    )
    documents = loader.load()

    python_splitter = RecursiveCharacterTextSplitter.from_language(
        language=Language.MULTILANGUAGE, chunk_size=2000, chunk_overlap=200
    )
    texts = python_splitter.split_documents(documents)

    embeddings = HuggingFaceEmbeddings(
        model_name="jinaai/jina-embeddings-v2-base-code",
        model_kwargs={'device': 'cuda' if torch.cuda.is_available() else 'cpu'},
    )

    vectordb = Chroma.from_documents(documents=texts, embedding=embeddings, persist_directory="chroma_db")
    retriever = vectordb.as_retriever(search_type="mmr", search_kwargs={"k": 8})

    return retriever

# Function for the GitHub Query page
def github_query_page():
    st.header("🔍 GitHub Query")
    st.markdown("Query your GitHub repositories and get detailed answers.")
    github_url = st.text_input("Enter the GitHub repository URL:")
    user_query = st.text_area("Enter your query:", height=300, placeholder="Enter your query here...")

    if st.button("Load Repository"):
        if github_url:
            retriever = process_github_repo(github_url)
            st.session_state['retriever'] = retriever
            st.session_state['repo_url'] = github_url
            st.success("Repository loaded successfully.")
        else:
            st.error("Please provide the GitHub URL.")

    if 'retriever' in st.session_state:
        if st.button("Run Query"):
            if user_query:
                qa = RetrievalQA.from_chain_type(
                    llm=llm,
                    chain_type="stuff",
                    retriever=st.session_state['retriever'],
                    verbose=True
                )
                query = f'''[INST] "<<SYS>>\n you are a helpful assistant you will be provided snippets of code and you should answer questions based on these code snippets. if you do not know the answer, just say you don't know. \n<</SYS>>\n\n"{user_query} based on the code provided?
                [/INST]'''
                result = qa.run(query)
                st.write("Query Result:")
                st.write(result)
            else:
                st.error("Please provide your query.")
    else:
        st.info("Load a repository first to run a query.")

def main():
    st.sidebar.title("Navigation")
    page = st.sidebar.radio("Go to", ["Home", "Document Files", "GitHub Query"])

    st.sidebar.markdown("""
    ### About
    This app allows you to:
    - Upload and analyze code files
    - Generate documentation and diagrams
    - Query GitHub repositories
    """)

    if page == "Home":
        home_page()
    elif page == "Document Files":
        document_files_page()
    elif page == "GitHub Query":
        github_query_page()


if __name__ == "__main__":
    main()
````
